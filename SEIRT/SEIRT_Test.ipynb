```python
import numpy as np
import matplotlib.pyplot as plt
from seirt_abm import ABMSEIRT
```

```python
cmap = {
        'S': (0, 0.5, 1),
        'E': (0.7, 0.5, 0),
        'I': (1, 0, 0),
        'R': (0, 0.9, 0.5),
        'T': (0.8, 0, 0.8)        
}
```

```python
am = ABMSEIRT(0.3/1000, 2, 3, 0.03, 1, 0, N=1000)
am.reset(10)
trajs = am.gillespie(1000, samples=10)
fig, ax = plt.subplots()

for (imax, times, traj) in trajs:
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 0, axis=1), c=cmap['S'], label='S')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 1, axis=1), c=cmap['E'], label='E')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 2, axis=1), c=cmap['I'], label='I')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 3, axis=1), c=cmap['R'], label='R')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 4, axis=1), c=cmap['T'], label='T')

ax.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from seirt_abm import ABMSEIRT
```

```python
am = ABMSEIRT(0.3/1000, 0.1, 0.1, 0.1, 0.1, 1., N=1000)
```

```python
am.reset(100)
```

```python
trajs = am.gillespie(1000, samples=10)
```

```python
fig, ax = plt.subplots()

for (imax, times, traj) in trajs:
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 0, axis=1), label='S')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 1, axis=1), label='E')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 2, axis=1), label='I')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 3, axis=1), label='R')
    ax.plot(times[:imax], np.sum(traj[:imax,:,0] == 4, axis=1), label='T')

ax.legend()
```
